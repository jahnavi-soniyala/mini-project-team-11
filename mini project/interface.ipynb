```python
def generate_amazon_links(ingredients):
    ingredient1, ingredient2, ingredient3=[i for i in ingredients]
    base_url = "https://www.amazon.in/s?k="
    link1 = f'<a href="{base_url}{ingredient1}" target="_blank">Search for {ingredient1} on Amazon</a>'
    link2 = f'<a href="{base_url}{ingredient2}" target="_blank">Search for {ingredient2} on Amazon</a>'
    link3 = f'<a href="{base_url}{ingredient3}" target="_blank">Search for {ingredient3} on Amazon</a>'
    
    return link1, link2, link3
```

```python
def result(input):
    df = pd.DataFrame([0]*len(x.columns), index=x.columns).T
    for i in input:
       df[i]=1
    d=loaded_model.predict(df)[0]
    link1,link2,link3 = generate_amazon_links(remedies[d]['herbs'])
    return f"you may had {d}\nherbs:{remedies[d]['herbs']}\npractices:{remedies[d]['practices']} \n", link1,link2,link3
```

```python
def solution(v, p, k):
    vata = """Vata Dosha Solutions:
Vata dosha is characterized by qualities such as light, dry, cold, rough, and mobile. When Vata is imbalanced, it can lead to issues such as anxiety, insomnia, dry skin, and digestive problems.

Diet:
- Eat warm, cooked, and nourishing foods.
- Include healthy fats like ghee and olive oil.
- Favor sweet, sour, and salty tastes.
- Avoid cold, dry, and raw foods.
- Stay hydrated with warm drinks.

Lifestyle:
- Maintain a regular routine with set meal and sleep times.
- Practice grounding activities like yoga, meditation, and walking.
- Avoid overstimulation and excessive travel.
- Use warm, oil-based massages to calm the nervous system.
- Ensure adequate rest and relaxation."""

    pitta = """Pitta Dosha Solutions:
Pitta dosha is associated with heat, intensity, and sharpness. Imbalance can cause anger, inflammation, heartburn, and skin rashes.

Diet:
- Eat cooling, hydrating foods such as cucumber, watermelon, and leafy greens.
- Favor sweet, bitter, and astringent tastes.
- Avoid spicy, oily, and fried foods.
- Drink plenty of water, and consider coconut water for extra cooling.
- Incorporate dairy, if tolerated, like milk and butter.

Lifestyle:
- Engage in calming activities like swimming or walking in nature.
- Practice cooling breathing exercises like sheetali pranayama.
- Avoid excessive heat and sun exposure.
- Maintain a balanced work-life schedule to prevent burnout.
- Use cooling oils like coconut oil for massages."""

    kapha = """Kapha Dosha Solutions:
Kapha dosha is characterized by heaviness, slowness, cold, and stability. Imbalance can result in weight gain, lethargy, congestion, and depression.

Diet:
- Eat light, warm, and spicy foods.
- Favor bitter, pungent, and astringent tastes.
- Avoid heavy, oily, and sweet foods.
- Drink warm beverages and limit dairy products.
- Incorporate plenty of vegetables and legumes.

Lifestyle:
- Engage in regular, vigorous exercise to boost metabolism.
- Practice stimulating activities like running or aerobics.
- Avoid daytime naps and maintain a regular sleep schedule.
- Use warming and invigorating oils for massages.
- Stay active and keep a dynamic lifestyle."""

    result = ""
    if v > p and v > k:
        result = vata
    elif p > v and p > k:
        result = pitta
    elif k > v and k > p:
        result = kapha
    elif v==p and p>k:
        print('you have vata-pitta dosha')
    elif v==k and k>p:
        print('you have vata-kapha dosha')
    elif k==p and p >v:
        print('you have pitta-kapha dosha')
    else:
        result = "You have a balanced dosha or a mixed dosha type. Consider a balanced approach combining solutions for all doshas."

    return result
```
